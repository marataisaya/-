```python
"""By this hand maded key words we can claster them into 14 groups"""
type_key_words = [
'Взнос',
'Возврат',
'Выдача',
'Зачисление Комиссия Корректировка Операции Поправки Увеличение Урегулирование Установление Безналичное Безналичный' ,
'Межфилиальные',
'Наличные',
'Оплата',
'Перевод',
'Плата',
'Платеж',
'Погашение',
'Покупка',
'Пополнение',
'Списание']
```

```python
types['type_cluster'] ='null'
```

```python
"""The loops for doing the clusterization"""
for k in range(14):
    for i in range(136):
        for j in types['type_description'].iloc[i].replace('.', '').replace('/', ' ').split(' '):
            for c in type_key_words[k].split(' '):
                if j == c:
                    types['type_cluster'].iloc[i] = k
```

### IV. Unsupervised learning. 
Do the Cluster analysis. Segment the customers. K-means,
Hierarchical Clustering. With different metrics, linkages. Visualize the clusters etc.
Look for the optimal number of the clusters

```python
fem_data
```

```python
###RFM score for fem data
# Converting to show date only
from datetime import datetime
fem_data["datetime"] = fem_data["datetime"].dt.date

# Creating variable that records recency
import datetime
snapshot_date = max(fem_data.datetime) + datetime.timedelta(days=1)

# Aggregate data by each client
clients = fem_data.groupby(['client_id']).agg({
    'datetime': lambda x: (snapshot_date - x.max()).days,
    'num': 'count',
    'sum': 'sum'})


# Renaming columns
clients.rename(columns = {'datetime': 'Recency',
                            'num': 'Frequency',
                            'sum': 'MonetaryValue'}, inplace=True)
```

```python
clients
```

```python
#to get positive values we use abs function
clients['MonetaryValue']=clients['MonetaryValue'].abs()
```

```python
clients
```

```python
#we transform the RFM table
from scipy import stats
clients_fix = pd.DataFrame()
clients_fix["Recency"] = stats.boxcox(clients['Recency'])[0]
clients_fix["Frequency"] = stats.boxcox(clients['Frequency'])[0]
clients_fix["MonetaryValue"] = pd.Series(np.cbrt(clients['MonetaryValue'])).values
clients_fix
```

```python
#each variable don’t have the same mean and variance => need to normalize
#to do this, we use StandardScaler object from scikit-learn library
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

scaler.fit(clients_fix)
clients_normal = scaler.transform(clients_fix)

print(clients_normal.mean(axis = 0))
print(clients_normal.std(axis = 0))
```

```python
print(clients_normal)
```

```python
#just showing it the in the dataframe
df = pd.DataFrame(clients_normal, columns = ['0','1','2'])
df
```

```python
#to make our maximum performance clustering, we need to determine which hyperparameter fits
#to determine which hyperparameter is the best, we can use the elbow method to find the number of cluster
from matplotlib import pyplot as plt
from sklearn.cluster import KMeans
sse = {}
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, random_state=42)
    kmeans.fit(clients_normal)
    sse[i] = kmeans.inertia_
plt.title('The Elbow Method')
plt.xlabel('i')
plt.ylabel('SSE')
sns.pointplot(x=list(sse.keys()), y=list(sse.values()))
plt.show()
```

```python
#the k-value=3 is the best hyperparameter because the next k-values tend to have a linear trend
# by fitting the model, we have clusters where each data belongs
model = KMeans(n_clusters=3, random_state=42)
model.fit(clients_normal)
model.labels_.shape
```

```python
#we can summarize the RFM table based on clusters and calculate the mean of each variable
clients["cluster"] = model.labels_
clients.groupby('cluster').agg({
    'Recency':'mean',
    'Frequency':'mean',
    'MonetaryValue':['mean', 'count']}).round(2)
```

```python
###Clusterasation for females
# creating the dataframe
df_cluster = pd.DataFrame(clients_normal, columns=['Recency', 'Frequency', 'MonetaryValue'])
df_cluster['id'] = clients.index
df_cluster['cluster'] = model.labels_

## melting the data
df_cluster_melt = pd.melt(df_cluster.reset_index(),
                      id_vars=['id', 'cluster'],
                      value_vars=['Recency','Frequency','MonetaryValue'],
                      var_name='Attribute',
                      value_name='Value')
df_cluster_melt.head()

# visualizing it
sns.lineplot('Attribute', 'Value', hue='cluster', data=df_cluster_melt)
```

By using this plot, we know how each segment differs.

We infer that cluster 0 is the most frequent, spend most, but they do not do transactions recently. Therefore, it could be the cluster of a loyal and old female clients.

Then, the cluster 1 is frequent averagely, less to spend averagely, but they do not do transactions recently. Therefore, it could be the cluster of old  and average class of female clients.

Finally, the cluster 2 is less frequent, less to spend, but they do transactions recently. Therefore, it could be the cluster of new female clients.
