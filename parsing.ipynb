```python
df = pd.DataFrame(columns=['BIN', '2019', '2020', '2021'])
counter = 0
```

```python
bin_list = bin_list[1:]
```

```python
# from webdriver_manager.chrome import ChromeDriverManager
from selenium.common.exceptions import WebDriverException
from selenium import webdriver
import pandas as pd
import requests

try:
    for i in range(len(bin_list)):
        driver = webdriver.Chrome('chromedriver.exe')
        # get geeksforgeeks.org
        driver.get("https://pk.uchet.kz")
        # get element

        element = driver.find_element_by_xpath("//input[@name='search']")

        # element2 = driver.find_element_by_class_name('btn-search primary').click()
        # send keys
        element.send_keys(bin_list[0])

        element2 = driver.find_element_by_class_name('primary').click()

        element3 = driver.find_element_by_class_name('company-link-open').click()

        element4 = driver.find_element_by_class_name('btn-block').click()

        login = driver.find_element_by_xpath("//input[@name='_username']")
        login.send_keys('alibek.nygmanov@zerde.gov.kz')

        password = driver.find_element_by_xpath("//input[@name='_password']")
        password.send_keys('Aa123456789')

        element5 = driver.find_element_by_class_name('btn-danger').click()

        table = driver.find_elements_by_xpath("//*[@class= 'table']")

        f = []
        for j in table:
            f.append(j.text.replace('\n',';').split(';'))
        aa = f[0]

        ###############
        b = bin_list[0]

        a = pd.DataFrame(aa)
        a['год']=a[0].str.split(':').str.get(0)
        a['налог']=a[0].str.split(':').str.get(1)
        del a[0]

        r2019 = (a[(a['год'] == '2019 год')]['налог']).tolist()
        if r2019 == []:
            nalog2019 = 0
        else:
            nalog2019 = int(r2019[0].replace(' ', '').replace('тг',''))

        r2020 = (a[(a['год'] == '2020 год')]['налог']).tolist()
        if r2020 == []:
            nalog2020 = 0
        else:
            nalog2020 = int(r2020[0].replace(' ', '').replace('тг',''))

        r2021 = (a[(a['год'] == '2021 год')]['налог']).tolist()
        if r2021 == []:
            nalog2021 = 0
        else:
            nalog2021 = int(r2021[0].replace(' ', '').replace('тг',''))

        df = df.append({'BIN': b, '2019': nalog2019, '2020': nalog2020, '2021': nalog2021}, ignore_index=True)

        counter+=1
        print(counter)
        print(bin_list[0])
        print('------------')
        bin_list.pop(0)
        driver.quit()
    
except WebDriverException:
    display(Javascript('IPython.notebook.execute_cell()'))
```

```python
df.to_csv('out.csv', index=False, sep=',')
```
